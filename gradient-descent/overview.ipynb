<h1> Gradient Descent </h1>

<h2> 2D example </h2>

```python
def sphere_2D(x):
    return x**2

def sphere_with_holes(x):
    return x**2 + 30*np.exp(-0.2*(x-2)**2) + 2*np.exp(-0.5*(x+5)**2)

def rasstrigin_2D(x):
    return x**2 - 22*np.cos(2*np.pi*x) + 20
```

```python
def min_fn(x):
    return sphere_2D(x)

x_min = -10
x_max = 10

def plot_curve():
    x_sp = np.linspace(x_min, x_max, 100)
    y_sp = min_fn(x_sp)
    plt.plot(x_sp, y_sp)

def sp(x_p):
    plot_curve()

    plt.scatter([x_p],min_fn(x_p), s=200)

def draw_line(x1, y1, x2, y2):
    plt.xlim(x_min, x_max)
    plt.ylim(-10, 100)

    plt.plot([x1, x2], [y1, y2], 'k-')


def text(vx, x, y):
    plt.text(x + 4, y + 10, f'vx: {np.round(vx, 2)}, x: {np.round(x, 2)}, y: {np.round(y, 2)}', fontsize=12, ha='right')
```

```python
def df(f, x):
    h = 0.0000001
    return ( f(x + h) - f(x) ) / h
```

```python
x_0 = 10
x = x_0
```

<h2> Change position by restarting the cell</h2>

```python
lr = 0.03

vx = df(min_fn, x)

x = x - lr * vx
y = min_fn(x)

sp(x)
draw_line(x, y, (x + vx), min_fn(x + vx))
text(vx, x, y)
```

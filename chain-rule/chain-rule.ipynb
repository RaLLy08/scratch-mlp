The goal is to find how Input values depend on Error on each layer. In order to decrease error.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define functions
def h(x):
    return x**2 + 1  # u = h(x)

def g(u):
    return 3 * u + 2  # v = g(u)

def f(v):
    return v**4      # y = f(v)

# Derivatives
def du_dx(x):
    return 2 * x

def dv_du(u):
    return 3 * np.ones_like(u)

def dy_dv(v):
    return 4 * v**3

# Composite function y = f(g(h(x)))
def y(x):
    return f(g(h(x)))

# Chain rule in Leibniz notation: dy/dx = dy/dv * dv/du * du/dx
def dy_dx(x):
    u = h(x)
    v = g(u)
    return dy_dv(v) * dv_du(u) * du_dx(x)

# x values
x = np.linspace(-3, 3, 400)
u = h(x)
v = g(u)
y_vals = y(x)
dy_dx_vals = dy_dx(x)

# Plotting with Leibniz notation labels
plt.figure(figsize=(12, 10))

# h(x): u
plt.subplot(2, 2, 1)
plt.plot(x, u, label="u = h(x) = x + 1")
plt.plot(x, du_dx(x), label=r"$\frac{du}{dx} = 2x$", linestyle='--')
plt.title("Step 1: $\\frac{du}{dx} = 2x$")
plt.xlabel("x")
plt.ylabel("u")
plt.grid(True)
plt.legend()

# g(u): v
plt.subplot(2, 2, 2)
plt.plot(x, v, label="v = g(u) = 3u + 2 = 3(x + 1) + 2")
plt.plot(x, dv_du(u), label=r"$\frac{dv}{du} = 3$", linestyle='--')
plt.title("Step 2: $\\frac{dv}{du} = 3$")
plt.xlabel("x")
plt.ylabel("v")
plt.grid(True)
plt.legend()

# f(v): y
plt.subplot(2, 2, 3)
plt.plot(x, y_vals, label="y = f(v) = v = (3x + 5)")
plt.plot(x, dy_dv(v), label=r"$\frac{dy}{dv} = 4v^3$", linestyle='--')
plt.title("Step 3: $\\frac{dy}{dv} = 4v^3$")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

# Final derivative
plt.subplot(2, 2, 4)
plt.plot(x, dy_dx_vals, label=r"$\frac{dy}{dx} = \frac{dy}{dv} \cdot \frac{dv}{du} \cdot \frac{du}{dx}$")
plt.title("Final Chain Rule Derivative")
plt.xlabel("x")
plt.ylabel("dy/dx")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
